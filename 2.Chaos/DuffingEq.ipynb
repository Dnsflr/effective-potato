```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
```

Duffing system function

```python
def duffing(t, xv):
    x, v = xv
    a, b, c = 1.0, 1.0, 0.2
    freq = 0.213
    omega = freq * 2 * np.pi
    f = 0.3  
    
    dxdt = v
    dvdt = b * x - a * x**3 - c * v + f * np.cos(omega * t)
    
    return [dxdt, dvdt]
```

Generating time vector 

```python
t_span = [0, 500]
num_points = 10000

t_eval = np.linspace(t_span[0], t_span[1], num_points)
```

Solving equatios

```python
y0 = [0.0, 0.15] 

solution = solve_ivp(duffing, t_span, y0, t_eval=t_eval)
```

Ekstraction of resoults

```python
x = solution.y[0]
v = solution.y[1]
t = solution.t
```

Plots of x(t) and v(t)

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True)
fig.suptitle(f'x(t) and v(t) plots for Duffing system', fontsize=16)

#x(t)
ax1.plot(t, x, label='x(t)', color='royalblue')
ax1.set_ylabel('x(t)')
ax1.legend(loc='upper right')
ax1.grid(True)

#v(t)
ax2.plot(t, v, label='v(t)', color='darkgreen')
ax2.set_xlabel('t')
ax2.set_ylabel('v(t)')
ax2.legend(loc='upper right')
ax2.grid(True)


plt.tight_layout() 
```
