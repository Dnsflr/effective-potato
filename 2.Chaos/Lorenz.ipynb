```python
from numpy import linspace
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
def lorenz(t, xyz):
    x, y , z = xyz
    s, r , b = 10, 28, 8/3
    return [s*(y - x), r*x - y - x*z, x*y - b*z]
```

```python
a, b = 0 , 40
t = linspace(a, b, 10000)


sol11 = solve_ivp(lorenz, [a, b], [1, 1, 1], t_eval=t)
sol12 = solve_ivp(lorenz, [a, b], [1, 1, 1.0000001], t_eval=t)
```

```python
plt.plot(sol11.y[0], sol11.y[2], color='blue', label='Initial condition: (1, 1, 1)')
plt.xlabel("$x$")
plt.ylabel("$z$")

ax = plt.gca()
ax.set_facecolor('black')
```

```python
plt.subplot(211)
plt.plot(sol11.t, sol11.y[0])
plt.xlabel("$t$")
plt.ylabel("$x_1(t)$")
plt.subplot(212)
plt.plot(sol11.t, sol11.y[0] - sol12.y[0])
plt.ylabel("$x_1(t) - x_2(t)$")
plt.xlabel("$t$")
```
