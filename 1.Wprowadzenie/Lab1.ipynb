```python
import numpy as np
import matplotlib.pyplot as plt
```

# Intro

Create an array of X values

```python
x = np.linspace(-np.pi, np.pi, 100)
```

Numpy math functions

```python
y1 = np.cos(x)
y2 = np.sin(x)
```

Plotting

```python
plt.plot(x, y1, 'r-', linewidth=1, label='cos(x)')
```

```python
plt.plot(x, y2, 'b--', linewidth=1,label='cos(x)')
```

Imporving the plots

define limits of x and y

```python
plt.xlim(x.min(), x.max())
plt.ylim(-1.2, 1.2)

plt.plot(x, y1, 'r-', linewidth=1, label='cos(x)')
plt.plot(x, y2, 'b--', linewidth=1,label='sin(x)')


plt.grid(alpha = 0.6)

plt.legend()
```

# Task 1

Creating x with 1000 points in the interval [-10, 10]

```python
x = np.linspace(-10, 10, 1000)
```

Plotting functions [sin(x), sin(2x), sin(x)*e^(-x^2/25)]

sin(x)

```python
y1 = np.sin(x)

plt.plot(x, y1, 'r-', linewidth=1, label='sin(x)')
```

sin(2x)

```python
y2 = np.sin(2*x)

plt.plot(x, y2, label='sin(2x)')
```

sin(x) * exp((-x^2)/25)

```python
y3 = np.sin(x) * np.exp((-x**2)/25)

plt.plot(x, y3, 'g--', label='sin(x)*e^(-x^2/25)')
```

all functions on one plot

```python
plt.plot(x, y1, 'r-', lw=1, label='sin(x)')
plt.plot(x, y2, linestyle='-',color='b',lw=2,label='sin(2x)')
plt.plot(x, y3, linestyle=':',color='g',lw=3, label='sin(x)*e^(-x^2/25)')
```

Legend and grid etc

```python
plt.plot(x, y1, 'r-', lw=1, label='sin(x)')
plt.plot(x, y2, linestyle='-',color='b',lw=2,label='sin(2x)')
plt.plot(x, y3, linestyle=':',color='g',lw=3, label='sin(x)*e^(-x^2/25)')


plt.title('Patterns')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.legend(loc='upper left', fontsize=8)

plt.savefig('patterns.png', dpi=300)
```

# Task 2

Gaussian distrubution mean = 15, sigma = 100

```python
x = np.random.normal(15, 100, 10000)
```

```python
plt.hist(x, bins=50, density=True, alpha=0.6, color='g')
```

```python
mean = np.mean(x)
std_dev = np.std(x)
```

```python
from matplotlib.lines import Line2D


plt.hist(x, bins=50, density=False, alpha=0.6, color='g')

legend_elements = [
    Line2D([0], [0], color='none', label=f'Mean: {round(mean,2)}'),
    Line2D([0], [0], color='none', label=f'Std: {round(std_dev,2)}')
]

plt.legend(handles=legend_elements, loc='upper right')

plt.title('Gaussian Distribution with Mean and Std Dev')
plt.ylabel(r'Density')
plt.xlabel(r'Value')
plt.show()
```

# Task 2'

Creating arrays

```python
n = 10000000
arr_of_num = []
P = 10
for i in range(P):
    x = np.random.uniform(-1, 1, n)
    arr_of_num.append(x)
    
```

summing arrays

```python
x_sum = np.sum(arr_of_num, axis=0)
```

rescaling

```python
x_new = np.mean(x_sum, axis=0) + np.std(x_sum, axis=0) * x_sum * np.sqrt(3. / P)
print(x_new)

plt.hist(x_new, bins=50, density=True, alpha=0.6, color='g')
plt.title('CTG Presentation')
```
